## Co-Relation (if two column is co-related with each other so we only use one column) mostly use when  relation is greatter then 85%

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(), annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(df, 0.8)
print(len(set(corr_features)))
print(corr_features)
```

```python
df2 = df.drop(labels=['target'],axis=1)
X = df2.iloc[:, :]
y = df['target'].iloc[:]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,
    y,
    test_size=0.2,
    random_state=20)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.transform(X_test)
```

```python
model_svm = svm.SVC()
model_svm.fit(X_train,y_train)
```

```python
# accuracy on training data
X_train_prediction = model_svm.predict(X_train)
training_data_accuracy = accuracy_score(y_train, X_train_prediction)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model_svm.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```
